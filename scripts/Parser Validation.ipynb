```python
import utility as ut
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
datapath_lr= '../OneDrive_1_11-10-2021/AD_P522R_F1_160/MRI_50um/mag_sos_wn.nii'
data_lr = ut.load_data_nii(datapath_lr)
datapath_hr= '../OneDrive_1_11-10-2021/AD_P522R_F1_160/MRI_25um/mag_sos_wn.nii'
data_hr = ut.load_data_nii(datapath_hr)
```

```python
## Crop and padd image
hr = data_hr[:,:,70]
lr_crop_pad_8 = ut.crop_pad_image_kspace(data_hr[:,:,70],pad=True,factor=8)
```

```python
plt.figure(figsize=(25,15))

error_image = ut.normalize_image(hr)- ut.normalize_image(lr_crop_pad_8)

plt.subplot(131),plt.imshow(ut.normalize_image(hr), cmap = 'gray',vmin=0,vmax=1)
plt.title('Original HR Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(error_image*10, cmap = 'gray',vmin=0,vmax=1)
plt.title('Error: Hr-cropped_padded fac 8'), plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(ut.normalize_image(lr_crop_pad_8), cmap = 'gray',vmin=0,vmax=1)
plt.title('Cropped-padded image fac 8'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
print(error_image.max())
print(error_image.min())
```
